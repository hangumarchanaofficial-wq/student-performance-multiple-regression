```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error , r2_score , mean_absolute_error

```

```python
df = pd.read_csv("../data/raw/Student_Performance.csv")
df.head()
```

```python
x = df[["Hours Studied","Previous Scores", "Sample Question Papers Practiced"]]
y = df["Performance Index"]
```

```python
X_train , X_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
regressor =LinearRegression()
```

```python
regressor.fit(X_train,y_train)
```

```python
y_pred_train = regressor.predict(X_train)
```

```python
residual = y_train - y_pred_train
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(y_pred_train, residual)
plt.axhline(y=0, linestyle="--")
plt.xlabel("Predicted Performance Index")
plt.ylabel("Residuals")
plt.title("Residuals vs Predicted Values (expanded Model)")
plt.show()
```

```python
print(f"r2Score : {r2_score(y_train, y_pred_train)}")
print(f"mseScore : {mean_squared_error(y_train, y_pred_train)}")
print(f"maeScore : {mean_absolute_error(y_train, y_pred_train)}")
```

Adding Sample Question Papers Practiced provides a small but consistent improvement in predictive performance, suggesting that targeted exam practice contributes incremental explanatory power beyond study time and prior ability.
